```python
import pandas as pd
from row64tools import ramdb
import paramiko
```

```python
df = pd.read_csv(r"C:\Users\mikha\Customers\BAH\Data\ais-2025-03-29.csv")
df1 = pd.read_csv(r"C:\Users\mikha\Customers\BAH\Data\ais-2025-03-30.csv")
df2= pd.read_csv(r"C:\Users\mikha\Customers\BAH\Data\ais-2025-03-31.csv")
```

```python
marine_traffic = pd.concat([df1, df1, df2], ignore_index=True)
marine_traffic = marine_traffic.sort_values(by='base_date_time')

marine_traffic['vessel_type'] = (marine_traffic['vessel_type'].fillna(0).replace([float('inf'),float('-inf')],0).astype(int).astype(str))
marine_traffic['cargo'] = (marine_traffic['cargo'].fillna(0).replace([float('inf'),float('-inf')],0).astype(int).astype(str))

```

```python
sample= marine_traffic.iloc[:100]
ramdb.save_from_df(sample, "C:\\Users\\mikha\\marine_traffic.ramdb")
```
