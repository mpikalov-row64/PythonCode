```python
import pandas as pd
from row64tools import ramdb

# Load CSV -> convert to ramdb
df = pd.read_csv(r"C:\Users\mikha\OneDrive\Documents\Demos\Supply Chain\warehouse_stream.csv")
ramdb.save_from_df(df, r"C:\Users\mikha\OneDrive\Documents\Demos\Supply Chain\Supply-Chain.ramdb")
print(f"Saved: {df.shape}")

# Now read it back (test round-trip)
df2 = ramdb.load_to_df(r"C:\Users\mikha\OneDrive\Documents\Demos\Supply Chain\Supply-Chain.ramdb")
print(f"Loaded: {df2.shape}")
print(df2.dtypes)

# Filter zone_telem like the scorer does
telem = df2[df2["record_type"] == "zone_telem"]
print(f"zone_telem rows: {len(telem)}")
print(telem[["zone_id","utilization","throughput","queue"]].head(3))
```
